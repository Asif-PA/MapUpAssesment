Series is a one-dimensional labeled array capable of holding any data type (integers, strings, floating point numbers, Python objects, etc.). The axis labels are collectively referred to as the index. The basic method to create a Series is to call:



Pandas Series
A Series is a one-dimensional array-like object that can hold various types of data, such as integers, floats, strings, and even other Python objects. It is similar to a column in an Excel spreadsheet or a single column in a database table.

Key Characteristics of a Series:

One-Dimensional: It holds a single column of data.
Indexed: Each element in a Series has an associated index, which can be numeric (default) or a custom index.
Homogeneous: All elements in a Series are of the same data type.

**DataFrame** is a 2-dimensional labeled data structure with columns of potentially different types. You can think of it like a spreadsheet or SQL table, or a dict of Series objects. It is generally the most commonly used pandas object

Along with the data, you can optionally pass index (row labels) and columns (column labels) arguments. If you pass an index and / or columns, you are guaranteeing the index and / or columns of the resulting DataFrame. Thus, a dict of Series plus a specific index will discard all data not matching up to the passed index.

If axis labels are not passed, they will be constructed from the input data based on common sense rules.

# Creating A Data Frame

From a Dictionary:

Dictionary of Lists: Each key represents a column name, and the corresponding list contains the data for that column.
Dictionary of Dictionaries: The outer dictionary keys represent the columns, and the inner dictionary keys represent the row indices.
Dictionary of Series: Similar to a dictionary of lists but uses pandas Series instead.


```python
# Dictionary of lists
data = {
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
}
df1 = pd.DataFrame(data)
df1

```

```python
# Dictionary of dictionaries
data = {
    'A': {0: 1, 1: 2, 2: 3},
    'B': {0: 4, 1: 5, 2: 6},
    'C': {0: 7, 1: 8, 2: 9}
}
df2 = pd.DataFrame(data)
df2
```

```python
# Dictionary of Series
data = {
    'A': pd.Series([1, 2, 3]),
    'B': pd.Series([4, 5, 6]),
    'C': pd.Series([7, 8, 9])
}
df3 = pd.DataFrame(data)
df3
```

```python
d = {
    "one": pd.Series([1.0, 2.0, 3.0], index=["a", "b", "c"]),
    "two": pd.Series([1.0, 2.0, 3.0, 4.0], index=["a", "b", "c", "d"]),
}
```

```python
c=pd.DataFrame(d)
```

```python
c
```

```python
e=pd.DataFrame(d, index=["d", "b", "f"])
```

```python
e
```

```python
#confusing
pd.DataFrame(d, index=["d", "b", "a"], columns=["two", "one"])
```
